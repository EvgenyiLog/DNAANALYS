```python
import cv2
def pngreader(filepath):
    image=cv2.imread(filepath,-1)
    #print(image.shape)
    print(image.dtype)
    h, w = image.shape[:2]
    print(f'Weigth= {w}')
    print(f'Heigth= {h}')
    assert image  is not None, "file could not be read, check with os.path.exists()"
    return image
import numpy as np
import matplotlib.pyplot as plt
def main():
    image=pngreader("C:/Users/evgen/Downloads/2712_1.png")
    # Specify parameters for the HOG descriptor

    # Pixel size in pixels (width, height). It must be smaller than the detection window size,
    # and must be chosen such that the resulting block size is smaller than the size of the detection window.
    cell_size = (3, 3)

    # Number of cells per block in each direction (x, y). must be selected for the result
    # Block size smaller than detection window
    num_cells_per_block = (2, 2)

    #   Block size in pixels (width, height). Must be an integer multiple of "cell size".
    # The block size must be smaller than the detection window.
    block_size = (num_cells_per_block[0] * cell_size[0],
              num_cells_per_block[1] * cell_size[1])

    #   Calculate the number of pixels that fit our image in the x and y directions
    x_cells = image.shape[1] // cell_size[0]
    y_cells = image.shape[0] // cell_size[1]

    # Horizontal distance between blocks, in cell size. must be an integer and must
    # Set it to (x_cells-num_cells_per_block[0]) / h_stride = integer.
    h_stride = 1

    # Vertical distance between blocks, in cell size. must be an integer and must
    # Set it to (y_cells - num_cells_per_block[1]) / v_stride = integer.
    v_stride = 1

    # Horizontal distance between blocks, in cell size. must be an integer and must
    # Set it to (x_cells-num_cells_per_block[0]) / h_stride = integer.
    h_stride = 1

    # Vertical distance between blocks, in cell size. must be an integer and must
    # Set it to (y_cells - num_cells_per_block[1]) / v_stride = integer.
    v_stride = 1

    # Block stride in pixels (horizontal, vertical). Must be an integer multiple of the pixel size.
    block_stride = (cell_size[0] * h_stride, cell_size[1] * v_stride)

    # Number of gradient orientation bins
    num_bins = 9
    
    # Specify the size of the detection window (region of interest) in pixels (width, height).
    # It must be an integer multiple of the "cell size" and must cover the entire image.
    # Since the detection window must be an integer multiple of the cell size, depending on your cell size,
    # Therefore, the generated detection window may be smaller than the image.
    # totally doable
    win_size = (x_cells * cell_size[0] , y_cells * cell_size[1])


    # Block stride in pixels (horizontal, vertical). Must be an integer multiple of the pixel size.# Set the parameters of the HOG descriptor using the variables defined above
    hog = cv2.HOGDescriptor(win_size, block_size, block_stride, cell_size, num_bins)

    # Compute the HOG descriptor of the grayscale image
    hog_descriptor = hog.compute(image)
    plt.show()


if __name__ == '__main__':
    main()
   
```
